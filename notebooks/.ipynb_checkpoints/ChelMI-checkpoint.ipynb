# Обработка данных кардиографа
Необходимо классифицировать данные кардиографа методом опорных векторов и найти точку наилучшей классификации. В качестве исходных данных будем использовать данные с кардиографа из ауд. 190а. 
Точность классификации будем определять, высчитывая чувствительность и специфичность. 

### Визуализация исходных данных
Для начала получим исходные данные и построим их график.

```python
import matplotlib.pyplot as pplt
with open ('data.txt', 'r') as f:
    data_list = [int(x) for x in f.read().splitlines()]
pplt.plot(data_list)
```

 Теперь аппрксимируем данные методом скользящего среднего и посмотрим на результат. Использование метода скользящего среднего приведет к потере некоторых значений, но с данном случае это несущественно.

```python
approx = []
s = 0
for i in range(4,len(data_list)-5):
    for j in range(i-4, i+5):
        s+=data_list[j]/10
    approx.append(s)
    s = 0
pplt.plot(approx)
```

### Классификация
Теперь применим к данным классификатор. Будем пользоваться методом опорных векторов. Для этого получим семейство прямых, для каждой найдем чувствительность и специфичность и среди них выберем лучшую.

Для определения точности классификации разделим данные на "больных" и "здоровых". Под чувствительностью будем понимать отношение количества "больных", классифицируемых как "больных", к реальному количеству больных. Под специфичностью будем понимать аналогичное отношение для "здоровых".

```python
sick = approx[358:426]
healthy = approx[:358]+approx[426:]
```

Заведем счетчики "больных" и "здоровых", а также определим начальное значение для опорного вектора.

```python
sick_count = healthy_count = 0
n = min(approx)
```

Запустим классификатор с шагом 10, а так же будем искать оптимальное значение $n$. Под оптимальным значением будем понимать максимально удаленую от начала координат точку на графике отношения чувствительностти к специфичности. Кроме того, будем сохранять значения чувствительности и специфичности для каждого $n$.

```python
saved_len = 0
saved_n = 0
sens = []
spec = []
while n<max(approx):
    for i in range(len(sick)):
        if sick[i] <= n:
            sick_count += 1
    for i in range(len(healthy)):
        if healthy[i] >= n:
            healthy_count += 1
    specifity = round(healthy_count*100/len(healthy))
    sensitivity = round(sick_count*100/len(sick))
    spec.append(specifity)
    sens.append(sensitivity)
    if (specifity ** 2 + sensitivity ** 2) ** 0.5 > saved_len:
        saved_len = (specifity ** 2 + sensitivity ** 2) ** 0.5
        saved_n = n
    n += 10
    sick_count = healthy_count = 0
```

Построим график отношения чувствительности к специфичности, а так же график исходных данных с наложенным опорным вектором.

```python
pplt.plot(sens, spec, c='g')
pplt.show()
print('n = ', saved_n)
pplt.axhline(saved_n, c='m')
pplt.plot(approx, c='r')
```
